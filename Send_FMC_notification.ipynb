```python
# ===============================================
#          FCM HTTP v1 Notification Sender
#          Author  : Kumar Saurabh
#          Version : v1.0
#          License : MIT
# ===============================================

# Install required packages
!pip install --quiet google-auth requests

# Import necessary libraries
import json
import requests
from google.oauth2 import service_account
from google.auth.transport.requests import Request

# ===========================
#   Configuration Section
# ===========================

# 1. Service Account JSON
# You can upload your service account JSON file to Colab and provide the path here,
# or paste the JSON content directly.

# Option 1: Provide path to uploaded JSON file
# SERVICE_ACCOUNT_FILE = './serviceAccountKey.json'  # Update this path

# Option 2: Paste JSON content directly
SERVICE_ACCOUNT_INFO = {
  "type": "",
  "project_id": "",
  "private_key_id": "",
  "private_key": "",
  "client_email": "",
  "client_id": "",
  "auth_uri": "",
  "token_uri": "",
  "auth_provider_x509_cert_url": "",
  "client_x509_cert_url": "",
  "universe_domain": ""
}

# 2. Firebase Project ID
PROJECT_ID = ""  # Update this with your project ID

# 3. Target Device FCM Token
TARGET_FCM_TOKEN = ''  # Update this with the target device's FCM token

# 4. Notification Details
NOTIFICATION_TITLE = 'Test Notification'
NOTIFICATION_BODY = 'This is a test message sent using FCM HTTP v1 API.'

# ===========================
#   Function Definitions
# ===========================

def get_access_token():
    """
    Generates an OAuth2 access token using the service account credentials.
    """
    try:
        # Option 1: From JSON file
        # credentials = service_account.Credentials.from_service_account_file(
        #     SERVICE_ACCOUNT_FILE,
        #     scopes=["https://www.googleapis.com/auth/firebase.messaging"]
        # )

        # Option 2: From JSON content
        credentials = service_account.Credentials.from_service_account_info(
            SERVICE_ACCOUNT_INFO,
            scopes=["https://www.googleapis.com/auth/firebase.messaging"]
        )

        request = Request()
        credentials.refresh(request)
        return credentials.token
    except Exception as e:
        print(f"Error obtaining access token: {e}")
        return None

def send_fcm_message(token, title, body):
    """
    Sends a notification message to a device using FCM HTTP v1 API.
    """
    access_token = get_access_token()
    if not access_token:
        print("Failed to obtain access token.")
        return

    url = f"https://fcm.googleapis.com/v1/projects/{PROJECT_ID}/messages:send"

    headers = {
        "Authorization": f"Bearer {access_token}",
        "Content-Type": "application/json; UTF-8",
    }

    message = {
        "message": {
            "token": token,
            "notification": {
                "title": title,
                "body": body
            }
            # You can add "data" payload here if needed
            # "data": {
            #     "key1": "value1",
            #     "key2": "value2"
            # }
        }
    }

    try:
        response = requests.post(url, headers=headers, json=message)
        if response.status_code == 200:
            print("Message sent successfully!")
            print("Response:", response.json())
        else:
            print(f"Failed to send message. Status code: {response.status_code}")
            print("Error:", response.json())
    except Exception as e:
        print(f"Error sending message: {e}")

# ===========================
#       Send Notification
# ===========================

send_fcm_message(TARGET_FCM_TOKEN, NOTIFICATION_TITLE, NOTIFICATION_BODY)
```
